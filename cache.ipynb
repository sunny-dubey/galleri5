```python
import torch
import gc
import os

print("ðŸ§¹ Clearing GPU VRAM...")

# Clean up Python objects but keep essential modules
keep_vars = {'torch', 'gc', 'os', 'keep_vars'}
for obj_name in list(globals().keys()):
    if obj_name not in keep_vars and not obj_name.startswith("_"):
        del globals()[obj_name]

# Run garbage collection
gc.collect()

# Clear CUDA cache if GPU is available
if torch.cuda.is_available():
    torch.cuda.empty_cache()
    torch.cuda.ipc_collect()
    print("âœ… Torch cache cleared")

# Show GPU status
os.system("nvidia-smi")
```
